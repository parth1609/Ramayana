```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.valmikiramayan.net/"
```

```python
# 2. Send an HTTP GET request to the URL
try:
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for bad status codes (4xx or 5xx)
    print(response)
except requests.exceptions.RequestException as e:
    print(f"Error fetching URL: {e}")
    exit()
```

```python
# 3. Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
frameset_tag = soup.find((['frame', 'iframe','frameset']))

if frameset_tag:
    print("Found the <frameset> tag:")
    print(frameset_tag)
else:
    print("No <frameset> tag found in the HTML.")
    # If no frameset, there are no frames to process
    exit()
```

```python
frame_tags = frameset_tag.find_all('frame')

if frame_tags:
    print("\nFound the <frame> tags:")
    for frame in frame_tags:
        print(frame)
else:
    print("\nNo <frame> tags found within the <frameset> tag.")
    # If no frames, there's nothing further to do in this context
    exit()
```

```python
# Find all <frame> tags
frame_tags = soup.find_all('frame')

target_frame_url = []
if frame_tags:
    print("Found the <frame> tags:")
    for frame in frame_tags:
        # Access the 'src' attribute using dictionary-like access
        target_frame_url.append(frame.get('src'))
        print(target_frame_url[0])
else:
    print("No <frame> tags found in the HTML.")
```

```python

all_href_links = []
# Find all 'a' tags with the 'href' attribute on the entire page
all_a_tags = soup.find_all('a', href=True)

if all_a_tags:
    for a_tag in all_a_tags:
        href_link = a_tag['href']
        all_href_links.append(href_link)
all_href_links
      
```

```python
all_kanda_sarga_links={}
def extract_sarga_links(soup, kanda_url):
    sarga_links = []
    all_a_tags = soup.find_all('a', href=True)  # Find all <a> tags with href

    if all_a_tags:  # Check if any <a> tags with href were found
        for a_tag in all_a_tags:  # Iterate through each <a> tag
            href_link = a_tag['href']  # Get the href attribute (the URL)
            title = a_tag.text.strip().lower()

            # Construct absolute URL
            if not href_link.startswith('http'):
                full_sarga_url = kanda_url.rsplit('/', 1)[0] + '/' + href_link
            else:
                full_sarga_url = href_link

            #  Filter condition: Check if the URL starts with the target domain
            if full_sarga_url.startswith('https://www.valmikiramayan.net/'):
                if "sarga" in title or "chapter" in title:
                    sarga_links.append({'title': a_tag.text.strip(), 'url': full_sarga_url})

    return sarga_links

    
```

```python

for kanda in kanda_links:
    kanda_name = kanda['name']
    kanda_url = kanda['url']
    print(f"\nFetching Sarga links for {kanda_name} from: {kanda_url}")
    try:
        response = requests.get(kanda_url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        sarga_links = extract_sarga_links(soup, kanda_url)  # Use the function
        all_kanda_sarga_links[kanda_name] = sarga_links
        print(f"Found {len(sarga_links)} Sarga links for {kanda_name}.")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching {kanda_url}: {e}")
    except Exception as e:
        print(f"An error occurred while processing {kanda_name}: {e}")

# Print the collected Sarga links for all Kandas
print("\nAll Kanda Sarga Links:")
for kanda, sarga_list in all_kanda_sarga_links.items():
    print(f"\n--- {kanda} ---")
    if sarga_list:
        for sarga in sarga_list:
            print(f"- {sarga['title']}: {sarga['url']}")
    else:
        print ("- No Sarga links found.")
```

```python

import re
import time # To add delays between requests

# Filter all_sarga_data for Bala Kanda only
bala_kanda_sargas = [sarga for sarga in all_sarga_data if sarga['kanda_name'] == 'Bala']

print(f"Found {len(bala_kanda_sargas)} Sargas for Bala Kanda.")



# --- Scraping Loop for Bala Kanda ---
for sarga_info in bala_kanda_sargas:
    url = sarga_info['url']
    kanda_name = sarga_info['kanda_name']
    sarga_index = sarga_info['index'] + 1 # Make it 1-based

    print(f"Scraping {kanda_name} - Sarga {sarga_index} from {url}...")

    try:
        response = requests.get(url)
        response.raise_for_status() # Check for HTTP errors
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all relevant paragraphs: shlokas, translations, etc.
        paragraphs = soup.find_all('p', class_=['SanSloka', 'pratipada', 'tat', 'verloc'])

        current_shloka_text = None
        current_shloka_num_str = None # To store the full '1-1-1' style number
        current_sarga_num = None # Extracted Sarga number

        for p in paragraphs:
            class_name = p.get('class')[0] if p.get('class') else None

            # --- Extract Sanskrit Shloka ---
            if class_name == 'SanSloka':
                # Ignore paragraphs that just contain audio controls
                if p.find('audio'):
                    continue

                shloka_raw_text = p.get_text(separator='\n', strip=True) # Use newline separator for <br>

                # Try to extract verse number (e.g., || 1-1-1)
                match = re.search(r'\|\|\s*(\d+)-(\d+)-(\d+)', shloka_raw_text)
                if match:
                    book_num, sarga_num_from_text, verse_num = match.groups()
                    current_sarga_num = int(sarga_num_from_text) # Update sarga number from text
                    current_shloka_num_str = f"{book_num}-{sarga_num_from_text}-{verse_num}"
                    # Clean the shloka text by removing the verse number part
                    current_shloka_text = shloka_raw_text.split('||')[0].strip()
                else:
                    current_shloka_text = shloka_raw_text.strip()
                  
            elif class_name == 'tat':
                if current_shloka_text: # Ensure we have a shloka to associate with
                    translation_text = p.get_text(strip=True)

                    # Clean up common artifacts like verse numbers in brackets
                    translation_text = re.sub(r'\[\d+-\d+-\d+\]', '', translation_text).strip()

                    # Add the extracted data
                    all_shloka_data.append({
                        "kanda_name": kanda_name,
                        "sarga": current_sarga_num if current_sarga_num else sarga_index, # Use extracted if available, else fallback
                        "shloka": current_shloka_text,
                        "translation": translation_text,
                        "url": url,
                        "verse_ref": current_shloka_num_str # Store the full reference if found
                    })

                    # Reset after capturing translation to avoid duplication
                    current_shloka_text = None
                    current_shloka_num_str = None
                    # current_sarga_num = None # Keep sarga num within the page loop

        # --- Polite delay ---
        time.sleep(0.5) # Wait half a second between requests

    except requests.exceptions.RequestException as e:
        print(f"  Error fetching {url}: {e}")
    except Exception as e:
        print(f"  An error occurred processing {url}: {e}")

print("\nScraping finished.")

```
